### For this demonstration, the 3-qubit W state will be used:
$$|W\rangle = \frac{1}{\sqrt{3}}|001\rangle + \frac{1}{\sqrt{3}}|010\rangle + \frac{1}{\sqrt{3}}|100\rangle$$

---

## Import MiQE

```python
from MiQE import *
```

## Initialise and build W state

```python
W = QuantumCircuit(num_qubits=3, structure='DensityMatrix')

W.gate(Ry(np.arccos(-1/3)), 0)
W.C(gate=H, control=0, target=1)
W.C(X, 1, 2)
W.C(X, 0, 1)
W.gate(X, 0)
```

## Convert to state vector

```python
W.to_state_vector(permanent=True)
```

## Display column vector (default)

```python
W.show()
```

## Display column vector (LaTeX)

```python
W.show('latex')
```

## Display in Dirac notation

```python
W.diracify()
```

## List and plot probability amplitudes

```python
W.plot_probs(output=['list', 'plot'])
```

## List and plot probability amplitudes by qubit

```python
W.plot_probs(output=['list', 'plot'], by_qubit=True)
```

## Measure individual qubits (non-destructive)

```python
W.measure(0, 1, collapse=False)
```

## Measure system (non-destructive)

```python
W.measure_all(collapse=False)
```

## Convert to density matrix

```python
W.to_density_matrix(permanent=True)
```

## Display density matrix (default)

```python
W.show()
```

## Display density matrix (LaTeX)

```python
W.show('latex')
```

## Dephase channel

```python
W.dephase(0, 1, 2, epsilon=0.8)
W.show('latex')
```

## Depolarise channel

```python
W.depolarise(1)
W.show('latex')
```

## Attempt to convert to state vector

```python
W.to_state_vector(permanent=True)
```

## Create noise simulation W state function

```python
def noisy_W_state():
    qc = QuantumCircuit(3, xyz_errors=[0.01, 0.01, 0.01], random_error=0.02)
    qc.gate(Ry(np.arccos(-1/3)), 0)
    qc.C(H, 0, 1)
    qc.C(X, 1, 2)
    qc.C(X, 0, 1)
    qc.gate(X, 0)
    return qc
```

## Execute noisy_W_state

```python
run_circuit(noisy_W_state, shots=1000, output=['list', 'plot'])
```
