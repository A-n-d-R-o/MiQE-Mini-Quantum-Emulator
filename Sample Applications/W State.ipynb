```python
from MiQE import *
```

```python
W_state = QuantumCircuit(3, structure='DensityMatrix')
W_state.gate(Ry(np.arccos(-1/3)), 0)
W_state.C(H, 0, 1)
W_state.C(X, 1, 2)
W_state.C(X, 0, 1)
W_state.gate(X, 0)
```

```python
W_state.show('latex')
```

```python
W_state.to_state_vector(True)
W_state.show('latex')
```

```python
W_state.diracify()
```
